```python
from rhf import RHF
import numpy as np
```

```python
np.random.seed(10007)
features = 30
rows = 1000
data = np.random.randint(100, size=(rows*features)).reshape(rows,features)
data
```

```python
type(data[0][0])
```

```python
rhf = RHF(max_height = 5, num_trees=100, check_duplicates=False, decremental=False, seed_state = 100007)
```

```python
%%time
rhf.fit(data)
```

```python
scores = rhf.get_scores()
```

```python
scores.sum()
```
